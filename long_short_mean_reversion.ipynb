```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller
from statsmodels.api import OLS
import yfinance as yf
from datetime import timedelta
```

```python
data = yf.Tickers("^GSPC AAPL TSLA AMZN GOOGL SBB-B.ST CAST.ST VOLV-B.ST 7974.T 7203.T")
history = data.history(start='2018-05-24', end='2021-09-01')
cl = history['Open']
cl = cl.drop('^GSPC', axis=1).interpolate('time').iloc[1:]
sp500 = history['Close']['^GSPC']
```

```python
cl_train = cl[:int(len(cl)/2)]
cl_test = cl[int(len(cl)/2)+1:]
```

```python
windows = [2,3,4]+[5*(i+1) for i in range(10)]
windows
```

```python
returns = []
cumrets = []
for window in windows:
    ret = cl_train.pct_change()
    rolling_ret = ret.rolling(window=window, axis=0).mean()

    weights = - (ret- rolling_ret)
    weights = weights.div(abs(weights).sum(axis=1), axis = 0)
    dailyret = (weights.shift(-1)*ret).sum(axis=1, skipna=False)
    
    returns.append(dailyret)
for dailyret in returns:
    cumret = dailyret.fillna(0).cumsum()
    cumret.plot()
    cumrets.append(cumret[-1])
plt.legend(windows)
#sp500.interpolate('time').pct_change().cumsum().plot()
```

```python
window= windows[np.argmax(cumrets)]
ret = cl_test.pct_change()
rolling_ret = ret.rolling(window=window, axis=0).mean()

weights = - (ret- rolling_ret)
weights = weights.div(abs(weights).sum(axis=1), axis = 0)
dailyret = (weights.shift(-1)*ret).sum(axis=1, skipna=False)

cumret = dailyret.fillna(0).cumsum()
cumret.plot()
```
