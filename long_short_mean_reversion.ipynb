```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller
from statsmodels.api import OLS
import yfinance as yf
from datetime import timedelta
```

```python
data = yf.Tickers("^GSPC AAPL TSLA AMZN GOOGL SBB-B.ST CAST.ST VOLV-B.ST 7974.T 7203.T")
history = data.history(start='2018-05-24', end='2021-09-01')
cl = history['Open']
cl = cl.drop('^GSPC', axis=1).interpolate('time').iloc[1:]
sp500 = history['Close']['^GSPC']
```

```python
windows = [i+2 for i in range(50)]
returns = []
cumrets = []
for window in windows:
    ret = cl.pct_change()
    rolling_ret = ret.rolling(window=window, axis=0).mean()

    weights = - (ret- rolling_ret)
    weights = weights.div(abs(weights).sum(axis=1), axis = 0)
    dailyret = (weights.shift(1)*ret).sum(axis=1, skipna=False)
    
    returns.append(dailyret)
for dailyret in returns:
    cumret = (dailyret + 1).cumprod() - 1
    cumret.plot()
    cumrets.append(cumret[-1])
plt.legend(windows)
```

```python
tickers = ""
symbols = pd.read_csv('bats_symbols.csv')['Name']
for symbol in symbols:
    tickers += " " +symbol
tickers = tickers[1:]
```

```python
data = yf.Tickers(tickers)
history = data.history(start='2017-04-04', end='2020-05-01')
```

```python
#Regular Ret
transaction_percentage = 0.015 #percentage is the transaction cost for the position
optimal_basket_size = 20#basket_sizes[np.argmax(cumrets)]
cl = history['Close']
ret = cl.pct_change()
rolling_ret = ret.rolling(window=3, axis=0).mean()
weights = - ret.add(- rolling_ret, axis=0)
weights = weights.mask(abs(weights).rank(axis=1, method='max', ascending=False) > optimal_basket_size, np.nan)
weights = weights.div(abs(weights).sum(axis=1), axis = 0)
dailyret = (weights.shift(1)*ret).sum(axis=1, skipna=True)
#dailyret.cumsum().plot()
CGR = (1 + dailyret).cumprod() - 1
CGR.plot()
```

```python
#market ret
transaction_percentage = 0.015 #percentage is the transaction cost for the position
optimal_basket_size = 20#basket_sizes[np.argmax(cumrets)]
cl = history['Close']
ret = cl.pct_change()
market_ret = ret.mean(axis=1)
weights = - ret.add(- market_ret, axis=0)
weights = weights.mask(abs(weights).rank(axis=1, method='max', ascending=False) > optimal_basket_size, np.nan)
weights = weights.div(abs(weights).sum(axis=1), axis = 0)
dailyret = (weights.shift(1)*ret).sum(axis=1, skipna=True)
#dailyret.cumsum().plot()
CGR = (1 + dailyret).cumprod() - 1
CGR.plot()
```

```python
start_date, end_date = dailyret.index[[0,-1]]
years = (end_date - start_date).days/365

CAGR = (CGR[-1])**(1/years) - 1
print(f'Cummalative Anual Growth Rate: {np.round(100*CAGR,2)}%')
```

```python
ret = sp500.pct_change()
rolling_ret = ret.rolling(window=2, axis=0).mean()
weights = - ret.add(- rolling_ret, axis=0)
weights = np.sign(weights)
dailyret = (weights.shift(1)*(ret))
CGR = (1 + dailyret).cumprod() - 1
CGR.plot()
```

```python
start_date, end_date = dailyret.index[[0,-1]]
years = (end_date - start_date).days/365

CAGR = (CGR[-1])**(1/years) - 1
print(f'Cummalative Anual Growth Rate: {np.round(100*CAGR,2)}%')
```
